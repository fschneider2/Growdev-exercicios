# 6) Crie um array bidimensional com numpy (com valores
pré-definidos) com M linhas e N colunas e em seguida crie um array
com N elementos e verifique se existe uma linha do primeiro array
que corresponde ao segundo array criado.

```python
from numpy.lib.twodim_base import mask_indices
array1 = np.array([[1,3,5,7,9],[2,4,6,8,10],[0,11,13,15,17],[12,14,16,18,20],
                   [0,1,3,9,19]])

array2 = np.array([1,3,5,7,9])
existe = False

c = ma.masked_equal(array1, array2)

for i in c.mask:
    if i.all():
        existe = True

print(existe)
```

11) Crie um programa para gerar a seguinte imagem em um array
numpy e exibir com opencv, a grossura dos círculos em preto são de
livre escolha, bem como as cores de fundo e do círculo.

```python
img = np.ones((300, 300, 3))

cor = (255,255,255)
tamanho = (60,80,100,120,140)
p = (150,150)

for i in tamanho:
    cv2.circle(img, (p), i, cor)

cv2_imshow(img)
```
